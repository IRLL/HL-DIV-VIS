```python
# Add column of epoch scores
for index, df in enumerate(dfList):
    temp = df
    fileLoc = pathList[index]
    print("fileLoc: " + str(fileLoc))
    temp['epoch_score'] = 0
    directory = os.path.join(tempTopDir,fileLoc)
    scores_path = os.path.join(directory, 'scores.txt')
    f = open(scores_path, 'r')
    data_txt = f.read()
    data_txt = data_txt.replace('[', '')
    data_txt = data_txt.replace(']', '')
    print("Got data")
    print(data_txt)
    epoch_scores = np.fromstring(data_txt, dtype=float, sep=',')
    indices = temp.index[temp['end_of_epoch']].tolist()
    indices.append(len(temp)-1)
    for state in indices:
        if ((state < len(temp)) and (state > 0)) :
            # if it is the end of an epoch, fill in the score
            if state in indices:
                list_index = indices.index(state)
                if (temp.iloc[state]['epoch_score'] == 0):
                    temp.iloc[state]['epoch_score'] = epoch_scores[list_index]
    dfList[index].update(temp)
```

```python
def find_distance(x1, y1, x2, y2):
    df = pd.DataFrame()
    # Set up series of x and y values
    df['x1'] = x1
#     print("df['x1']")
#     print(df['x1'])
    df['y1'] = y1
#     print("df['y1']")
#     print(df['y1'])
    
#     print("x2")
#     print(x2)
    df['x2'] = x2
    df['x2'] = df['x2'].shift(-1)
    df['x2'].iloc[-1] = 0
#     print("df['x2']")
#     print(df['x2'])
    
#     print("y2")
#     print(y2)
    df['y2'] = y2
    df['y2'] = df['y2'].shift(-1)
    df['y2'].iloc[-1] = 0
#     print("df['y2']")
#     print(df['y2'])
    
#     print("Diffs:")
#     print(df.iloc[0,3])
    # Get difference of squared values
    df['x_tmp_diff'] = (df['x2']-df['x1'])
#     print("df['x_tmp_diff']")
#     print(df['x_tmp_diff'])
    df['x_tmp_sq'] = (df['x_tmp_diff'])**2
#     print("df['x_tmp_sq']")
#     print(df['x_tmp_sq'])
    df['y_tmp'] = (df['y2']-df['y1'])**2
#     print("df['y_tmp']")
#     print(df['y_tmp'])
    df['d_tmp'] = (df['x_tmp_sq']) + (df['y_tmp'])
#     print("df['d_tmp']")
#     print(df['d_tmp'])
    
    # Take square root
    df['d'] = np.sqrt(df['d_tmp'])
#     df['d'] = df['d'].shift(1)
#     df['d'].iloc[0] = 0
    return df['d']
```
